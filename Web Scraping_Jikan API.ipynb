```python
manga_df_list = []
error_count = 0
error_id_list = []
counter = 0
batch_size = 500
current_batch_number = 123

# Total 61195
# Start from 17281 next time
for id in miss_li:
    try:
        r = requests.get(f'https://api.jikan.moe/v4/manga/{id}/full')
        rate_limit_counter = 0
        while r.status_code == 429:
                print(f'Rate limit exceeded at id {id}. Retrying in 1 second...')
                rate_limit_counter += 1
                time.sleep(1)
                if rate_limit_counter > 60:
                     print('Rate limit exceeded for 60 times, exit the program.')
                     raise Exception('Rate limit exceeds 60 times')
                continue
        df = pd.DataFrame(r.json()).T
        manga_df_list.append(df)
        counter += 1

        # Save to CSV every 500 entries
        if counter % batch_size == 0:
            manga = pd.concat(manga_df_list)
            manga = manga.set_index('mal_id')
            manga.to_csv(f'manga_batch_{current_batch_number + 1}.csv')
            manga_df_list = []  # Reset list after saving
            current_batch_number += 1

    except Exception as e:
        print(r.status_code)
        print(e)
        print(f'Error happens at id {id}')
        error_count += 1
        error_id_list.append(id)
        continue

# Save any remaining entries after the loop
if manga_df_list:
    manga = pd.concat(manga_df_list)
    manga = manga.set_index('mal_id')
    manga.to_csv(f'manga_batch_{current_batch_number + 2}.csv')

with open('error_list_miss_v2.txt', 'a') as f:
    f.writelines([str(id)+'\n' for id in error_id_list])
```
